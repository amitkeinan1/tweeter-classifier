```python
import pickle
import pandas as pd
```

```python
test = pd.read_excel("data//test.xlsx")
```

```python
predictions = pickle.load(open("test_predictions.pkl", "rb"))
important_words = pickle.load(open("test_important_words.pkl", "rb"))
```

```python
users = test["users"]
sides = test["sides"]
test
```

```python
labels_dict = {"L": "שמאל", "R": "ימין"}
labels = [labels_dict[side] for side in sides]

predictions_dict = labels_dict = {0: "שמאל"
                                  , 1: "ימין"}
predictions = [predictions_dict[pred] for pred in predictions]
```

```python
names = ["נועם שוסטר",
        "מיקי חיימוביץ'",
        "ארז תדמור",
        "צבי האוזר",
        "אבי ניסנקורן",
        "גלעד ארדן",
        "בועז גולן",
        "נפולת",
        "בנימין נתניהו",
        "חן ליברמן"]
```

```python
results_df = pd.DataFrame({"users": users, "names": names, "labels": labels, "predictions": predictions, "words helped decide": important_words})
results_df
```
